```python
df_crescimento = df_completo.groupby(['ano', 'bioma'])['desmatado'].sum().reset_index()
```

```python
df_crescimento
```

```python
df_crescimento['desmatado_milhoes'] = (df_crescimento['desmatado'] / 1000000).round(2)
```

```python
df_crescimento
```

```python
plt.figure(figsize=(12, 6))
```

```python
grafico1 = sns.lineplot(data = df_crescimento, x='ano', y='desmatado_milhoes', hue='bioma', marker='o', linewidth=2.5)
plt.title('Evolução do Desmatamento Total por Bioma (2000-2023)', fontsize=16, pad=20)
plt.xlabel('Ano', fontsize=12)
plt.ylabel('Área Desmatada em Milhões (km²)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Bioma', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.gca().yaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))
plt.tight_layout()
plt.show()
```

```python
df_estados = df_completo.groupby('sigla_uf')[['desmatado', 'area_total']].sum().reset_index()
```

```python
df_estados['desmatado_milhoes'] = (df_estados['desmatado'] / 1000000).round(2)
```

```python
df_estados.sort_values('desmatado_milhoes', ascending=False)
```

```python
df_estados['areatotal_milhoes'] = (df_estados['area_total'] / 1000000).round(2)
```

```python
df_estados.sort_values('areatotal_milhoes', ascending=False)
```

```python
df_estados['porcentagem'] = (df_estados['desmatado'] / df_estados['area_total']) * 100
```

```python
df_estados.sort_values('porcentagem', ascending=False)
```

```python
df_absoluto = df_estados.sort_values('desmatado_milhoes', ascending=False)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot( data= df_absoluto, x='desmatado_milhoes', y='sigla_uf', hue='desmatado_milhoes', legend=False)

plt.title('Ranking Absoluto: Área Total Desmatada (km²)', fontsize=14)
plt.xlabel('Área Desmatada em Milhoes (km²)')
plt.ylabel('Estado')
plt.grid(True, axis='x', linestyle='--', alpha=0.3)
plt.gca().xaxis.set_major_formatter(ticker.FormatStrFormatter('%.2f'))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
df_relativo = df_estados.sort_values('porcentagem', ascending=False)
sns.barplot(data=df_relativo, x='porcentagem', y='sigla_uf', hue='porcentagem', legend=False)

plt.title('Ranking Relativo Desmatado (%)', fontsize=14)
plt.xlabel('Porcentagem (%)')
plt.ylabel('Estado')
plt.grid(True, axis='x', linestyle='--', alpha=0.3)
plt.show()
```
